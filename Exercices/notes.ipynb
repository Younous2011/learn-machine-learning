# Chapitre 1 : Structure matricièle

```python
import pandas as pd
import numpy as np
import random
```

```python
def random_notes():
    return random.randint(0, 20)
```

```python
# On définit les prénoms des élèves
prenoms = ["Caroline", "Youssef", "Micheal", "Badr", "Chady", "Ismael", "Eymen", "Aslam", "Saleh", "Soufiane", "Sanaa"]
prenomsY = prenoms + ["Younous"]
# On définit les matières
matieres = ["Maths", "Français", "Histoire-géo", "Physique-Chimie", "SVT", "Anglais"]
# On initialise la liste des notes de tous les élèves
notes = []

# On itère sur tous les élèves
for i  in range(11): 
    # On initialise une liste vide pour chaque élève
    notes_eleve = []
    # Pour chaque élève, on itère sur toutes les matières
    for j in range(6):
        # Pour chaque élève et pour chaque matière, on rajoute une notye aléatoire à la liste des notes d'un élève
        notes_eleve.append(random_notes())
    # On rajoute la liste des notes globales
    notes.append(notes_eleve)
# On rajoute les notes de Younous
notes.append([18, 20, 20, 20, 20, 100])
# On affiche les notes
print(notes)
```

```python
df = pd.DataFrame(notes, columns=matieres, index=prenomsY)
```

```python
df
```

```python
df.loc["Younous", "Maths"] = 20
df
```

```python
df["Total"] = df["Maths"] + df["Français"] + df["Histoire-géo"] + df["Physique-Chimie"] + df["SVT"] + df["Anglais"]
```

```python
df["Moyenne"] = df["Total"] / 6
```

```python
df[matieres]
```

```python
df["mean"] = df[matieres].mean(axis=1)
```

```python
df
```

```python
maskSup = df["mean"] >= 10
df[maskSup]
```

```python
maskInf = df["mean"] < 10
df[maskInf]
```

# Chapitre 2 : Structure en tuple

```python
"""
    notes = [
        ["Caroline", "Maths", 18],
        ["Caroline", "Français", 12],
        ...,
        ["Sanaa", "Anglais", 12] 
    ]
"""
```

```python
notes = []
for prenom in prenoms:
    for matiere in matieres:
        notes.append([prenom, matiere, random_notes()])

notes.append(["Younous", "Maths", 21])
notes.append(["Younous", "Français", 19])
notes.append(["Younous", "Histoire-géo", 19.5])
notes.append(["Younous", "Physique-Chimie", 17.75])
notes.append(["Younous", "SVT", 19])
notes.append(["Younous", "Anglais", 19.99])
notes
```

```python
df = pd.DataFrame(notes, columns=["Prenom", "Matière", "Note"])
df.head()
```

```python
mask = df["Prenom"] == "Caroline"
mask
```

```python
df_caroline = df[mask]
df_caroline
```

```python
df_caroline["Note"].mean()
```

```python
df.groupby("Prenom").mean()
```

```python
df.groupby("Prenom", as_index=False).agg({"Note": "mean"})
```

```python
df.groupby("Matière", as_index=False).agg({"Note": "mean"})
```

```python
df.groupby("Matière", as_index=False).agg({"Note": "max"})
```

```python
df.groupby("Matière", as_index=False).agg({"Note": "min"})
```

```python
df.groupby("Prenom", as_index=False).agg({"Note": "max"})
```

```python
df.groupby("Matière", as_index=False)["Note"].transform(max)
```

```python
df_m = df.copy()
df_m["NoteMaxMatière"] = df_m.groupby("Matière", as_index=False)["Note"].transform(max)
mask = df_m["Note"] == df_m["NoteMaxMatière"]
df_m[mask]
```

```python
df_p = df.copy()
df_p["NoteMaxEleve"] = df_p.groupby("Prenom", as_index=False)["Note"].transform(max)
mask = df_p["Note"] == df_p["NoteMaxEleve"]
df_p[mask]
```

```python
df_t = df.copy()
df_t["NoteMinEleve"] = df_t.groupby("Prenom", as_index=False)["Note"].transform(min)
mask = df_t["Note"] == df_t["NoteMinEleve"]
df_t[mask]
```

```python
df_pivot = df.pivot_table(index="Prenom", columns="Matière", values="Note")
df_pivot
```
